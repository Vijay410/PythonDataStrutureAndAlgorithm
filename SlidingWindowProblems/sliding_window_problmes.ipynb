#### 1Ô∏è‚É£ Maximum sum of subarray of size k

###### Approch 1. Brute Force Method

```python
# PROBLEM 1: Maximum Sum of Subarray of Size k
# Goal: Given an array and window size k, find the maximum sum of any contiguous subarray of size k.
# Example:
# arr = [12, -1, -7, 8, -16, 30, 16, 28]
# k = 3

# Brute Force Approach
arr = [12, -1, -7, 8, -16, 30, 16, 28]
k = 3

# Iterate over the array and ensure the window does not exceed the array boundary
max_sum = 0
for i in range(len(arr) - k + 1):      # starting index of the window
    current_sum = 0

    # Calculate the sum of the current window
    for j in range(i, i + k):          # elements inside the window
        current_sum += arr[j]

    max_sum = max(current_sum, max_sum)

print(max_sum)

# Explanation:
# This brute-force approach uses nested loops, resulting in a Time Complexity of O(n^2).
# Therefore, it becomes inefficient for large datasets.
```

##### Using Sliding window approch

```python
# Find the maximum sum of a subarray of size k using Sliding Window

arr = [12, -1, -7, 8, -16, 30, 16, 28]
k = 3

# Step 1: Calculate the initial window sum (first k elements)
ws = sum(arr[:k])            # Initial window: [12, -1, -7]
max_sum = ws

# Step 2: Slide the window from index k to the end of the array
for i in range(k, len(arr)):  # Start from k because the first window is already summed
    # Update the window sum:
    # Add the new element coming into the window ‚Üí arr[i]
    # Subtract the old element leaving the window ‚Üí arr[i - k]
    ws += arr[i] - arr[i - k]

    # Update maximum sum found so far
    max_sum = max(ws, max_sum)

print(max_sum)
```

#### 2Ô∏è‚É£ Sum of each window of size k

2Ô∏è‚É£ Sum of each window of size k

üìò Problem Statement

Given an array and window size k,
return the sum of every contiguous subarray of size k.

```python
# -----------------------------
# Problem:
# Find the sum of every contiguous subarray (window) of size k
# -----------------------------

# Input
arr = [12, -1, -7, 8, -16, 30, 16, 28]
k = 3

# Expected Output
# [4, 0, -15, 22, 30, 74]

# ============================================================
# Brute Force Approach
# Time Complexity: O(n * k)
# Space Complexity: O(1) (excluding output list)
# ============================================================

res = []

# Outer loop selects the starting index of each window
for i in range(len(arr) - k + 1):
    current_sum = 0

    # Inner loop calculates the sum of the current window
    for j in range(i, i + k):
        current_sum += arr[j]

    # Store sum of the current window
    res.append(current_sum)

print("Brute Force Result:", res)


# ============================================================
# Sliding Window Approach (Optimized)
# Time Complexity: O(n)
# Space Complexity: O(1) (excluding output list)
# ============================================================

res = []

# Step 1: Calculate sum of the first window
window_sum = sum(arr[:k])
res.append(window_sum)

# Step 2: Slide the window across the array
for i in range(k, len(arr)):

    # Add the new element entering the window (right side)
    # Remove the old element leaving the window (left side)
    window_sum += arr[i] - arr[i - k]

    # Store the updated window sum
    res.append(window_sum)

print("Sliding Window Result:", res)
```

### 3Ô∏è‚É£ Average of subarrays of size k

#Input
arr = [12, -1, -7, 8, -16, 30, 16, 28]
k = 3


```python
# -------------------------------------------------
# Problem:
# Find the average of every contiguous subarray
# (sliding window) of size k
# -------------------------------------------------

# Input
arr = [12, -1, -7, 8, -16, 30, 16, 28]
k = 3

# Expected Output
# [1.33, 0.00, -5.00, 7.33, 10.00, 24.67]

# =================================================
# Brute Force Approach
# Time Complexity: O(n * k)
# Space Complexity: O(1) (excluding output list)
# =================================================

res = []

# Outer loop chooses the starting index of each window
for i in range(len(arr) - k + 1):
    window_sum = 0

    # Inner loop calculates the sum of the current window
    for j in range(i, i + k):
        window_sum += arr[j]

    # Compute average of the current window (rounded to 2 decimals)
    avg_val = round(window_sum / k, 2)

    # Store result
    res.append(avg_val)

print("Brute Force Result:", res)


# =================================================
# Sliding Window Approach (Optimized)
# Time Complexity: O(n)
# Space Complexity: O(1) (excluding output list)
# =================================================

res = []

# Step 1: Compute the sum of the first window
window_sum = sum(arr[:k])
res.append(round(window_sum / k, 2))

# Step 2: Slide the window one element at a time
for i in range(k, len(arr)):

    # Add new element entering the window
    # Remove old element leaving the window
    window_sum += arr[i] - arr[i - k]

    # Calculate and store average of the current window
    res.append(round(window_sum / k, 2))

print("Sliding Window Result:", res)
```
