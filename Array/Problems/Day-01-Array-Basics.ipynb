```python
# Find Maximum value
arr = [10,12,13,14]
max_val =  arr[0]
for i in arr:
    if i > max_val:
        max_val = i
print(f"Maximum value in array: {max_val}")
```

```python
# Find Minimum value
arr = [10,12,13,14]
max_val =  arr[0]
for i in arr:
    if i < max_val:
        max_val = i
print(f"Minimum value in array: {max_val}")
```

```python
# Sum of elements in array
arr = [10,12,13,14]
total_sum = 0
for i in arr:
    total_sum += i
print(f"sum of all elemtnts: {total_sum}")
```
